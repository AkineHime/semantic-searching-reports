```python
from sklearn.datasets import load_files
import pandas as pd

# Load the dataset
newsgroups_data = load_files(r'..\data\20_newsgroups', encoding='latin1')

# Create a DataFrame
df = pd.DataFrame({'text': newsgroups_data.data, 'target': newsgroups_data.target})

# Map target names
df['target_name'] = df['target'].apply(lambda x: newsgroups_data.target_names[x])

df.head()
```
